```python
au = ['rougher.input.feed_au',
      'rougher.output.concentrate_au',
      'primary_cleaner.output.concentrate_au',
      'final.output.concentrate_au']
pb = ['rougher.input.feed_pb',
      'rougher.output.concentrate_pb',
      'primary_cleaner.output.concentrate_pb',
      'final.output.concentrate_pb']
ag = ['rougher.input.feed_ag',
      'rougher.output.concentrate_ag',
      'primary_cleaner.output.concentrate_ag',
      'final.output.concentrate_ag']




def plot_concentrations(df_train, au, pb, ag):
    fig, axs = plt.subplots(1, 3, figsize=(15,5))
    for i, metal in enumerate(['Au', 'Pb', 'Ag']):
        axs[i].bar(['Feed', 'Rougher', 'Primary Cleaner', 'Final'], 
                   [df_train[col].mean() for col in eval(metal.lower())])
        axs[i].set_title(metal)
        axs[i].set_xlabel('Stage')
        axs[i].set_ylabel('Concentration')

plot_concentrations(df_train, au, pb, ag)
```

```python
name_dict={
    0:'Концентрация золота',
    1:'Концентрация свинца',
    2:'Концентрация серебра'
}

def plot(df, column_lists):
    for idx, columns in enumerate(column_lists):
        plt.figure(figsize=(10,6))
        for col in columns:
            plt.hist(df[col], alpha=0.5, bins=20, label=col)
        plt.legend()
        plt.xlabel("Концентрация")
        plt.ylabel("Частота")
        plt.title(name_dict.get(idx, f'Plot {idx}'))
        plt.show()
plot(df_train, [au, pb, ag])
```

### 3.1. Напишите функцию для вычисления итоговой sMAPE.

Формула SMAPE выглядит следующим образом: \
**SMAPE = (1/n) * Σ(|прогноз – фактический| / ((|фактический| + |прогноз|)/2) * 100**

Напишем функцию, принимающую на вход: \
`a` - фактические значения; \
`f` - прогнозируемое значение.

```python
def smape( a , f ):
    return 1 / len(a) * np.sum(2 * np.abs(f - a) / (np.abs(a) + np.abs(f)) * 100)
def w_smape (score1, score2):
    return (-score1.mean()*0.25) + (-score2.mean()*0.75)
```

```python
# стандартизация непрерывных переменных
X_train_rougher = df_train_rougher.drop(['rougher.output.recovery'], axis=1)
X_test_rougher = df_test_rougher.drop(['rougher.output.recovery'], axis=1)
X_train_final = df_train_final.drop(['final.output.recovery'], axis=1)
X_test_final = df_test_final.drop(['final.output.recovery'], axis=1)

scaler_rougher = StandardScaler()
scaler_rougher.fit(X_train_rougher)
X_train_rougher_scaled = scaler_rougher.transform(X_train_rougher)
X_test_rougher_scaled = scaler_rougher.transform(X_test_rougher) 

scaler_final = StandardScaler()
scaler_final.fit(X_train_final)
X_train_final_scaled = scaler_final.transform(X_train_final)
X_test_final_scaled = scaler_final.transform(X_test_final)
```

Занесём нашу функцию в `make_scorer` для дальнейшей передачи в `cross_val_score`.

```python
smape_scorer = make_scorer(smape, greater_is_better=False)
```

### 3.2. Обучите разные модели и оцените их качество кросс-валидацией.

Попробуем обучить и подобрать гиперпараметры у `LinearRegression` и `RandomForestRegressor`.\
Сначала сделаем это для `df_train_rougher`, а затем для `df_train_final`.

```python
state = 1
y_train_rougher = df_train_rougher['rougher.output.recovery']

model_lr = LinearRegression()
scores_lr_train_rougher = cross_val_score(model_lr, 
                                          X_train_rougher_scaled, 
                                          y_train_rougher,
                                          cv=5, 
                                          scoring=smape_scorer)
print('Средняя оценка качества линейной регрессии -', -scores_lr_train_rougher.mean())
```

Займемся поиском лучших `n_estimators`, `max_depth` и `min_samples_leaf` для случайного леса.

```python
#поиск гиперпараметров для случайного леса на тренировочной выборке для rougher
depths = [None] + [i for i in range(2, 11)]
best_result = 0
best_params = {}

for est in range(20, 100, 20):
    for depth in depths:
        for leaf in range(1, 2):
            model_rf = RandomForestRegressor(random_state=state,
                                             n_estimators=est,
                                             max_depth=depth,
                                             min_samples_leaf=leaf) 
            scores = cross_val_score(model_rf,
                                     X_train_rougher_scaled, 
                                     y_train_rougher, 
                                     cv=5, 
                                     scoring=smape_scorer) 
            mean_score = scores.mean()
            if mean_score < best_result:
                best_result = mean_score
                best_params = {'n_estimators': est, 
                               'max_depth': depth, 
                               'min_samples_leaf': leaf}

print("Наилучшее значение SMAPE:", -best_result)
print("Наилучшие параметры:", best_params)
```

```python
# проверяем + включаем sMAPE в свою переменную
model_rf_rougher = RandomForestRegressor(random_state=state,
                                             n_estimators=20,
                                             max_depth=2,
                                             min_samples_leaf=1)
scores_rf_train_rougher = cross_val_score(model_rf, 
                                          X_train_rougher_scaled, 
                                          y_train_rougher, 
                                          cv=5, 
                                          scoring=smape_scorer)
print('Средняя оценка качества линейной регрессии -', -scores_rf_train_rougher.mean())
```

Пока что линейная регрессия показывает лучший результат. Проверим данные модели на выборке с целевым параметром `final.output.recovery`.

```python
y_train_final = df_train_final['final.output.recovery']

model_lr = LinearRegression()
scores_lr_train_final = cross_val_score(model_lr, X_train_final_scaled, y_train_final, cv=5, scoring=smape_scorer)
print('Средняя оценка SMAPE линейной регрессии -', -scores_lr_train_final.mean())
```

```python
#поиск гиперпараметров для случайного леса на тренировочной выборке для final

best_result = 0
best_params = {}

for est in range(20, 100, 20):
    for depth in depths:
        for leaf in range(1, 2):
            model_rf = RandomForestRegressor(random_state=state,
                                             n_estimators=est,
                                             max_depth=depth,
                                             min_samples_leaf=leaf) 
            scores = cross_val_score(model_rf, X_train_final_scaled, y_train_final, cv=5, scoring=smape_scorer) 
            mean_score = scores.mean()
            if mean_score < best_result:
                best_result = mean_score
                best_params = {'n_estimators': est, 
                               'max_depth': depth, 
                               'min_samples_leaf': leaf}

print("Наилучшее значение SMAPE:", -best_result)
print("Наилучшие параметры:", best_params)
```

```python
# проверяем + включаем sMAPE в свою переменную
model_rf_final = RandomForestRegressor(random_state=state,
                                             n_estimators=20,
                                             max_depth=None,
                                             min_samples_leaf=1)
scores_rf_train_final = cross_val_score(model_rf, X_train_final_scaled, y_train_final, cv=5, scoring=smape_scorer)
print('Средняя оценка качества линейной регрессии -', -scores_rf_train_final.mean())
```

Тут победа за случайным лесом. Оценим итоговый sMAPE на обучающей выборке.

```python
print('Итоговый sMAPE линейной регрессии:', w_smape(scores_lr_train_rougher, scores_lr_train_final))
print('Итоговый sMAPE случайного леса:', w_smape(scores_rf_train_rougher, scores_rf_train_final))
```

По результатам оценки итогового sMAPE **случайный лес** оказывается более релевантным. Его и будем пропускать через тестовую выборку.
